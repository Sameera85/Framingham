### 1. Data Decription:
<a id=section1></a>
This dataset contains information related to cardiovascular disease risk factors for a group of individuals. Cardiovascular diseases, including heart disease and stroke, are significant health concerns worldwide. Understanding the risk factors associated with these diseases is essential for prevention and management.

In the early 1900s, heart diseases has long been the #1 cause of death in the United States. The Framingham Heart Study, launched in 1948, is a long-term research project that was developed to identify risk factors for cardiovascular disease.

The project was named after Framingham, a town in Massachusetts chosen as the study site. The initial cohort included over 5,200 Framingham residents aged 30 to 62 without heart disease symptoms. Regular medical tests and lifestyle data were collected to observe disease development and connections to heart health over time.

   ##### 1.1.1 description of the columns in the dataset.

- **male (Categorical: 0 or 1):**:
    - 0: Female
    - 1: Male
- **age (Numeric - Integer):** Age of the individual in years.
- **education (Categorical: 1, 2, 3, 4):**:
    - 1: Some High School
    - 2: High School or GED
    - 3: Some College or Vocational School
    - 4: College
- **currentSmoker (Categorical: 0 or 1):**:
    - 0: No
    - 1: Yes
- **cigsPerDay (Numeric - Integer):** Number of cigarettes smoked per day (for smokers).
- **BPMeds (Categorical: 0, 1, or 'Missing'):**:
    - 0: Not taking blood pressure medication
    - 1: Taking blood pressure medication
    - 'Missing': Missing data
- **prevalentStroke (Categorical: 0 or 1):**:
    - 0: No history of stroke
    - 1: History of stroke
- **prevalentHyp (Categorical: 0 or 1):**:
    - 0: No prevalent hypertension (high blood pressure)
    - 1: Prevalent hypertension
- **diabetes (Categorical: 0 or 1):**:
    - 0: No diabetes
    - 1: Diabetes
- **totChol (Numeric - Float):** Total cholesterol level in the blood.
- **sysBP (Numeric - Float):** Systolic blood pressure, the top number in a blood pressure reading.
- **diaBP (Numeric - Float):** Diastolic blood pressure, the bottom number in a blood pressure reading.
- **BMI (Numeric - Float):** Body Mass Index, a measure of body weight relative to height.
- **heartRate (Numeric - Float):** Heart rate, the number of heartbeats per minute.
- **glucose (Numeric - Float):** Blood glucose level, a measure of sugar in the blood.
- **TenYearCHD (Categorical: 0 or 1):**:
    - 0: Lower risk of coronary heart disease (CHD) in the next 10 years
    - 1: Higher risk of coronary heart disease (CHD) in the next 10 years

- From the above output we can see that **education**, **cigsPerDay**, **BPMeds**, __totChol__, **BMI**, **heartRate** and **glucose**  columns contains __maximum null values__. We will see how to deal with them.


```python
# Impute missing values in 'education' with the mode (most frequent value)
df['education'].fillna(df['education'].mode()[0], inplace=True)

# Impute missing values in 'cigsPerDay' with the median
df['cigsPerDay'].fillna(df['cigsPerDay'].median(), inplace=True)


# Fill missing values in 'BPMeds' with a new category, e.g., 'Missing'
df['BPMeds'] = df['BPMeds'].cat.add_categories('Missing')
df['BPMeds'].fillna('Missing', inplace=True)


# Impute missing values in 'totChol' with the median
df['totChol'].fillna(df['totChol'].median(), inplace=True)

# Impute missing values in 'BMI' with the median
df['BMI'].fillna(df['BMI'].median(), inplace=True)

# Impute missing values in 'heartRate' with the median
df['heartRate'].fillna(df['heartRate'].median(), inplace=True)

# Impute missing values in 'glucose' with the median
df['glucose'].fillna(df['glucose'].median(), inplace=True)

# Print the updated DataFrame with missing values filled
df.isnull().sum()  # Verify that there are no more missing values
```

### 4.2 "What is the distribution of key cardiovascular risk factors in the dataset, and how does their skewness impact the understanding of the data's central tendency and spread? 


<a id=section402></a>

```python
# Define colors for each variable
colors = ['orange', 'green', 'red', 'purple', 'blue', 'grey']

# Create subplots for each variable
plt.figure(figsize=(23, 15))
plt.subplots_adjust(wspace=0.3, hspace=0.3)

variables = ['glucose', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate']

# Define skewness labels based on your criteria
def get_skew_label(skewness):
    if skewness < -1 or skewness > 1:
        return 'Highly Skewed'
    elif (-1 <= skewness <= -0.5) or (0.5 <= skewness <= 1):
        return 'Moderately Skewed'
    else:
        return 'Approximately Symmetric'

for i, var in enumerate(variables, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[var], color=colors[i - 1], kde=True)  # Plot histogram and KDE
    plt.title(f'Distribution of {var}')
    
    # Calculate skewness
    skewness = df[var].skew()
    skew_label = get_skew_label(skewness)
    
    plt.text(0.75, 0.9, f'Skewness: {skewness:.2f}\n({skew_label})', transform=plt.gca().transAxes, fontsize=10)

plt.show()
```

- Observations :

  - We can see Glucose, Total Cholesterol, Systolic BP, BMI, Diastolic BP & Heart Rate is Right Skewed.

### 4.3 "How does the distribution of heart rate and age groups in the dataset provide insights into the demographics and health characteristics of the surveyed population, and is there any noticeable correlation between these factors? 


<a id=section403></a>

```python
# Define age and heart rate grouping functions using lambda functions
age_groups = lambda x: 0 if x <= 40 else (1 if x <= 55 else 2)
heart_rate_groups = lambda x: 0 if x <= 60 else (1 if x <= 100 else 2)

# Apply the grouping functions and create new columns
df['age_groups'] = df['age'].apply(age_groups)
df['heart_rate_groups'] = df['heartRate'].apply(heart_rate_groups)
```

```python
df.head()
```

#### Age Groups:
- Age <= 40 are Adults.
- Age > 40 and <= 55 are Middle-Aged.
- Above 55 are Senior.
#### HeartRate Groups:

- HeartRate <= 60 is Low
- HeartRate between 60 to 100 is Normal.
- HeartRate above 100 is High.

```python
# Create a DataFrame for Plotly
plotly_df = df.copy()
plotly_df['heart_rate_groups'] = plotly_df['heart_rate_groups'].map({0: 'Low', 1: 'Normal', 2: 'High'})
plotly_df['age_groups'] = plotly_df['age_groups'].map({0: 'Adults', 1: 'Middle-Aged', 2: 'Senior'})

# Create subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=('HeartRate (Grouped)', 'Count by Age Group'))

# Create the first plot (HeartRate Grouped)
fig.add_trace(go.Bar(x=plotly_df['heart_rate_groups'].value_counts().index, y=plotly_df['heart_rate_groups'].value_counts(), marker_color='lightcoral'), row=1, col=1)
fig.update_xaxes(title_text='Heart Rate Group', row=1, col=1)
fig.update_yaxes(title_text='Count', row=1, col=1)

# Create the second plot (Count by Age Group)
fig.add_trace(go.Bar(x=plotly_df['age_groups'].value_counts().index, y=plotly_df['age_groups'].value_counts(), marker_color='lightblue'), row=1, col=2)
fig.update_xaxes(title_text='Age Group', row=1, col=2)
fig.update_yaxes(title_text='Count', row=1, col=2)

# Update layout and show the plot
fig.update_layout(height=400, width=1000, showlegend=False)
fig.update_layout(title_text="Heart Rate and Age Group Analysis")
fig.show()
```

- Observations :

   - We observe the majority of subjects exhibit a normal heart rate, followed by those with resting or low heart rates, and a smaller number with high heart rates.
   - Additionally, we observe a higher number of middle-aged adults in our data, followed by seniors, and then young adults."

### 4.4 "How do systolic and diastolic blood pressures vary across different age groups and genders, and are there any noticeable patterns or differences that could indicate potential health trends or risk factors? 


<a id=section404></a>

```python
# Set the figure size and adjust subplots with extra vertical space
plt.figure(figsize=(23, 12))
plt.subplots_adjust(wspace=0.2, hspace=0.5)  # Adjust hspace to add extra space

# Define age group labels
age_group_labels = ['Adults', 'Middle-Aged', 'Senior']

# Create the first set of boxenplots for Sys. BP
plt.subplot(2, 2, 1)
sns.boxenplot(x='age_groups', y='sysBP', hue='male', data=df, palette='rainbow')
plt.xlabel('Age Group / Gender')
plt.ylabel('Sys. BP')
plt.xticks(ticks=[0, 1, 2], labels=age_group_labels)
plt.title('Sys. BP by Age Group & Gender')
plt.legend(title='Gender')

# Create the second set of boxenplots for Dia. BP
plt.subplot(2, 2, 2)
sns.boxenplot(x='age_groups', y='diaBP', hue='male', data=df, palette='pastel')
plt.xlabel('Age Group / Gender')
plt.ylabel('Dia. BP')
plt.xticks(ticks=[0, 1, 2], labels=age_group_labels)
plt.title('Dia. BP by Age Group & Gender')
plt.legend(title='Gender')

# Create the first set of barplots for Sys. BP
plt.subplot(2, 2, 3)
sns.barplot(x='age_groups', y='sysBP', hue='male', data=df, palette='rainbow')
plt.xlabel('Age Group')
plt.ylabel('Sys. BP')
plt.title('Sys. BP by Age Group & Gender')
plt.legend(title='Gender')

# Create the second set of barplots for Dia. BP
plt.subplot(2, 2, 4)
sns.barplot(x='age_groups', y='diaBP', hue='male', data=df, palette='pastel')
plt.xlabel('Age Group')
plt.ylabel('Dia. BP')
plt.title('Dia. BP by Age Group & Gender')
plt.legend(title='Gender')

plt.show()
```

- Observations : 
  - in general, both systolic blood pressure (SBP) and diastolic blood pressure (DBP) tend to increase with age. However, the rate of increase is different for SBP and DBP. SBP tends to increase more rapidly than DBP, especially after the age of 50.

### 4.6 Is there a difference in the number of male and female patients with coronary heart disease? 


<a id=section406></a>

```python
# Create a count plot
sns.set_style("whitegrid")  # Add grid lines
plot = sns.catplot(x='male', hue='TenYearCHD', data=df, kind='count', palette='Paired', height=6, legend=False)

# Modify the axis labels
plt.xlabel('Gender', fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'], fontsize=12)
plt.ylabel('No. of Patients', fontsize=14)
plt.yticks(fontsize=12)

# Add a legend
plt.legend(['Neg.', 'Pos.'], title='CHD Status', title_fontsize=12, fontsize=12)

# Add a title
plt.title('Coronary Heart Disease (CHD) by Gender', fontsize=16)

# Calculate and display the percentage and count on the count plot
total_counts = len(df)
for p in plot.ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total_counts)
    count = int(p.get_height())
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    plot.ax.annotate(f'{percentage}\nCount: {count}', (x, y), ha='center', va='bottom', fontsize=12)

# Adjust plot aesthetics
sns.despine(left=True)
plot.ax.set_axisbelow(True)  # Move grid lines behind the bars

# Show the plot
plt.show()
```

- Observations :

  - The bivariate analysis plot illustrates the presence or absence of Chronic Heart Disease (CHD) based on gender.

  - Observations reveal that a significant portion of the population is not affected by CHD.

  - On the negative side, approximately 80 to 90% of females do not have CHD, while around 60 to 70% of males fall into the same category.

  - On the positive side, only about 10% of both females and males are suffering from CHD.

  - These findings indicate that our dataset exhibits an imbalance, with roughly 80 to 90% of cases categorized as negative (no CHD), and about 10 to 15% classified as positive (CHD).

### 4.8 How does the prevalence of diabetes vary across different age groups, and what percentage of people in each age group have diabetes? 


<a id=section408></a>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a count plot for 'diabetes' by 'age_groups'
plt.figure(figsize=(10, 7))

ax = sns.countplot(x='age_groups', hue='diabetes', data=df, palette='rainbow')
plt.xlabel('Age Group')
plt.ylabel('No. of Patients')
plt.xticks(ticks=[0, 1, 2], labels=['Adults', 'Middle-Aged', 'Senior'])
plt.legend(title='Diabetes', labels=['Negative', 'Positive'])
plt.title('Diabetes by Age Group')

# Calculate and display percentages above the bars
total_count = len(df)

for p in ax.patches:
    height = p.get_height()
    percentage = height / total_count * 100
    ax.text(p.get_x() + p.get_width() / 2, height + 5, f'{percentage:.1f}%', ha='center')

plt.show()
```

- The number of people with a negative diabetes diagnosis is significantly higher than the number of people with a positive diabetes diagnosis in all age groups, thouths there is no apparent correlation between a person's age and their diabetes diagnosis

### 4.9 How are systolic blood pressure, diastolic blood pressure, total cholesterol, and 10-year coronary heart disease risk related to each other?



<a id=section409></a>

```python
# Create a pairplot
sns.pairplot(data=df, vars=['sysBP', 'diaBP', 'totChol'], hue='TenYearCHD', markers=["o", "s"])
plt.suptitle('Pairplot of sysBP, diaBP, totChol vs. TenYearCHD', y=1.02)
plt.show()
```

 - The Pairplot shows a number of interesting observations. For example:

   - There is a positive correlation between sysBP and TenYearCHD, meaning that people with higher systolic blood pressure have a higher risk of developing coronary heart disease within 10 years.
   - There is a positive correlation between diaBP and TenYearCHD, meaning that people with higher diastolic blood pressure also have a higher risk of developing coronary heart disease within 10 years.
   - There is a positive correlation between totChol and TenYearCHD, meaning that people with higher total cholesterol levels also have a higher risk of developing coronary heart disease within 10 years.
   - There is a positive correlation between sysBP and diaBP, meaning that people with higher systolic blood pressure tend to also have higher diastolic blood pressure.
   - There is a positive correlation between sysBP and totChol, meaning that people with higher systolic blood pressure tend to also have higher total cholesterol levels.
   - There is a positive correlation between diaBP and totChol, meaning that people with higher diastolic blood pressure tend to also have higher total cholesterol levels.
   
- These observations suggest that all four of these variables are important risk factors for coronary heart disease. It is important to note that the Pairplot only shows correlations, and correlation does not equal causation. However, the correlations observed in the Pairplot are consistent with the results of many other studies that have shown that sysBP, diaBP, totChol, and TenYearCHD are all risk factors for coronary heart disease.

- The Framingham Heart Study dataset offers valuable insights into cardiovascular risk factors and provides a foundation for further analysis and predictive modeling. This analysis has revealed several key findings, including the impact of age, gender, and risk factors on coronary heart disease risk. Further analysis and modeling can help develop tools for risk prediction and prevention.
